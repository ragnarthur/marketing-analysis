###  - Tratamento de valores Ausentes
- tratando valores ausentes na coluna "Income" (substituindo por mediana ou média)

#### Porquê escolher a mediana?!?
- A mediana não é afetada por valores extremos (outliers). Em datasets de renda, é comum haver valores muito altos ou muito baixos que podem distorcer a média. A mediana, sendo o valor central, representa melhor a tendência central dos dados sem ser influenciada por esses extremos.

- Renda é frequentemente uma variável com distribuição assimétrica, onde a maioria das pessoas ganha valores próximos ao centro da distribuição, mas há algumas pessoas que ganham muito mais. Em tais casos, a mediana fornece uma melhor representação da renda típica do que a média.

- Preencher valores ausentes com a mediana é uma abordagem simples e eficaz que garante que a distribuição dos dados não seja distorcida significativamente, mantendo a integridade das análises subsequentes.

##### Comparação com Outras Abordagens
- Média:
    - A média pode ser influenciada por valores atipicamente altos ou baixos, o que pode não representar bem a tendência central dos dados, especialmente em distribuições assimétricas como a de renda.

- Moda:
    - A moda pode ser útil se a distribuição for unimodal e os dados forem categóricos ou discretos, mas para variáveis contínuas como renda, a moda pode não ser tão representativa ou até inexistente se todos os valores forem únicos.

- Modelos de Inputação:
    - Técnicas mais sofisticadas como KNN imputation ou modelos de regressão podem ser usados para preencher valores ausentes, mas são mais complexos e computacionalmente intensivos. Para muitas análises exploratórias, a simplicidade da mediana é suficiente e apropriada.

```python
# Visualizar a distribuição da coluna Income
plt.figure(figsize=(10,6))
sns.histplot(df['Income'].dropna(), kde=True)
plt.title("Distribuição de Income")
plt.xlabel("Income")
plt.ylabel("Frequência")
plt.show()
```

##### Se a distribuição for assimétrica, com uma cauda longa, a escolha da mediana é confirmada como uma abordagem sensata. Se a distribuição fosse normal, a média poderia ser uma escolha igualmente válida.

```python
# preenchendo os valores ausentes na coluna Icome com a mediana
df['Income'] = df['Income'].fillna(df['Income'].median())
```

##  - Análise de Segmentação 
- Podemos criar segmentos de clientes com base nos índices RFM. Uma abordagem simples é dividir os clientes em quartis para cada índice e depois combinar esses segmentos. Vamos começar segmentando cada índice em quartis

```python
# Listar as colunas do DataFrame para verificar se estão corretas
print("Colunas do DataFrame:", df.columns.tolist())

# Preencher valores ausentes na coluna Income com a mediana
df['Income'] = df['Income'].fillna(df['Income'].median())

# Calcular o índice Frequency
df['Frequency'] = df[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)

# Calcular o índice Monetary
df['Monetary'] = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)

# Exibir estatísticas descritivas para a coluna Monetary
monetary_statistics = df['Monetary'].describe()
print("Estatísticas descritivas de Monetary:", monetary_statistics)

# Segmentar Recency, Frequency e Monetary em quartis
df['Recency_Segment'] = pd.qcut(df['Recency'], 4, labels=[4, 3, 2, 1])
df['Frequency_Segment'] = pd.qcut(df['Frequency'], 4, labels=[1, 2, 3, 4])
df['Monetary_Segment'] = pd.qcut(df['Monetary'], 4, labels=[1, 2, 3, 4])

# Calcular o valor RFM total
df['RFM_Score'] = df['Recency_Segment'].astype(int) + df['Frequency_Segment'].astype(int) + df['Monetary_Segment'].astype(int)

# Contar a distribuição dos scores RFM
rfm_counts = df['RFM_Score'].value_counts().sort_index()

# Visualizar a distribuição dos scores RFM
plt.figure(figsize=(10, 6))
sns.barplot(x=rfm_counts.index, y=rfm_counts.values, hue=rfm_counts.index, palette='viridis', dodge=False)
plt.legend([],[], frameon=False)
plt.title('Distribuição dos Scores RFM')
plt.xlabel('RFM Score')
plt.ylabel('Contagem')
plt.show()

# Analisar estatísticas descritivas dos segmentos RFM
rfm_segments_analysis = df.groupby('RFM_Score').agg({
    'Recency': ['mean', 'median'],
    'Frequency': ['mean', 'median'],
    'Monetary': ['mean', 'median'],
    'ID': 'count'
}).rename(columns={'ID': 'Count'})

rfm_segments_analysis
```

#### Análise dos Segmentos RFM
- **RFM Score 3**: Clientes menos recentes (média de 86 dias), menor frequência (média de 5.5 compras) e menor valor monetário (média de 39.3).

- **RFM Score 12**: Clientes mais recentes (média de 12 dias), maior frequência (média de 26.5 compras) e maior valor monetário (média de 1443).
    - A partir da análise dos segmentos, podemos identificar grupos de clientes valiosos (RFM Scores mais altos) e menos valiosos (RFM Scores mais baixos). Essa informação pode ser usada para direcionar campanhas de marketing mais eficazes.

##  - Análise de Camapanhas de Marketing

```python
# Verificar a resposta às campanhas de marketing
campaigns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']

# Adicionar a coluna 'TotalAccepted' que soma todas as campanhas aceitas
df['TotalAccepted'] = df[campaigns].sum(axis=1)

# Analisar a resposta às campanhas por segmento RFM
campaign_response = df.groupby('RFM_Score')['TotalAccepted'].mean().reset_index()

# Visualizar a resposta média às campanhas por segmento RFM
plt.figure(figsize=(10, 6))
sns.barplot(data=campaign_response, x='RFM_Score', y='TotalAccepted', palette='viridis')
plt.title('Resposta Média às Campanhas por Segmento RFM')
plt.xlabel('RFM Score')
plt.ylabel('Média de Campanhas Aceitas')
plt.show()

campaign_response
```

#### Análise da Resposta às Campanhas de Marketing
- Clientes com RFM Scores mais altos (9-12) tendem a aceitar mais campanhas de marketing, com médias de campanhas aceitas variando de 0.47 a 0.80.

- Clientes com RFM Scores mais baixos (3-6) têm uma resposta significativamente menor, com médias de campanhas aceitas variando de 0.05 a 0.10.

### Conclusão 
- **Clientes** Valiosos: Foi identificado que clientes com RFM Scores mais altos são mais valiosos e mais receptivos às campanhas de marketing.

- **Segmentação**: A segmentação por RFM Scores nos ajudou a entender diferentes grupos de clientes e suas respostas às campanhas de marketing.

- **Estratégia de Marketing**: Com base nesses insights, podemos direcionar campanhas mais intensas para clientes de RFM Scores mais altos e desenvolver estratégias de reativação para clientes com RFM Scores mais baixos.
